```
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```
pcla_2020 = pd.read_csv('PCLA_2020')
pcla_2019 = pd.read_csv('PCLA_2019')
```

```
pcla_2019.head()
```

```
newdf19 = pcla_2019.drop(columns=['spamReport', 'spamReportPercent','sentFromGid', 'bounced', 'bouncedPercent', 'category', 'emailName','deliveredPercent', 'unsubscribesPercent','sentCount', 'initialClicks', 'initialClicksPercent', 'conversionsPercent'])
newdf19.head()
```

```
total_emails_19 = newdf19['subjectLine'].count()
print(total_emails)
members_19 = 1575
```

```
#Unique Open Rate 19 
total_opens_19 = newdf19['opens'].sum()
total_unique_opens_19 = newdf19['uniqueOpens'].sum()
open_rate_19 = round((total_unique_opens_19 / total_opens_19) * 100)
print(total_opens_19)
print(total_unique_opens_19)
print(open_rate_19)
```

```
#Open Rate 19 
total_unique_19 = newdf19['uniqueOpens'].sum()
total_delivered_19 = newdf19['delivered'].sum()
open_rate_imodules_19 = round((total_unique_19 / total_delivered_19) * 100)
print(open_rate_imodules_19)
```

```
#Unique Click Rate 19
total_clicks_19 = newdf19['clicks'].sum()
total_unique_clicks_19 = newdf19['uniqueClicks'].sum()
click_rate_19 = round((total_unique_clicks_19 / total_clicks_19) * 100)
print(total_clicks_19)
print(total_unique_clicks_19)
print(click_rate_19)
```

```
#Conversions to Emails Sent 19
conversions_19 = newdf19['conversions'].sum()
print(conversions_19)
```

```
#Unscribes 19 
unscribes_19 = newdf19['unsubscribes'].sum()
print(unscribes_19)
```

```
total_emails_20 = newdf20['subjectLine'].count()
print(total_emails)
members_19 = 1522
```

```
#Unique open Rate 20 
total_opens_20 = newdf20['opens'].sum()
total_unique_opens_20 = newdf20['uniqueOpens'].sum()
open_rate_20 = round((total_unique_opens_20 / total_opens_20) * 100)
print(total_opens_20)
print(total_unique_opens_20)
print(open_rate_20)
```

```
#Open Rate 20
total_unique_20 = newdf20['uniqueOpens'].sum()
total_delivered_20 = newdf20['delivered'].sum()
open_rate_imodules_20 = round((total_unique_20 / total_delivered_20) * 100)
print(open_rate_imodules_20)
```

```
#Unique Click Rate 20
total_clicks_20 = newdf20['clicks'].sum()
total_unique_clicks_20 = newdf20['uniqueClicks'].sum()
click_rate_20 = round((total_unique_clicks_20 / total_clicks_20) * 100)
print(total_clicks_20)
print(total_unique_clicks_20)
print(click_rate_20)
```

```
#Conversions to Emails Sent 20
conversions_20 = newdf20['conversions'].sum()
print(conversions_20)
```

```
#Unscribes 20
unscribes_20 = newdf20['unsubscribes'].sum()
print(unscribes_20)
```

```
# 2019 vs 2020

#Unique Open Rate Change
open_change_unique = ((open_rate_20 - open_rate_19) / open_rate_19) * 100
print(open_change_unique)
#Open Rate iModules Change
open_change_rate = ((open_rate_imodules_20 - open_rate_imodules_19) / open_rate_imodules_19) * 100
print(open_change_rate)
#Click Rate Change
click_change = ((click_rate_20 - click_rate_19) / click_rate_19) * 100
print(click_change)
#Conversions Change
conversion_change = ((conversions_20 - conversions_19) / conversions_19) * 100
print(conversion_change)
```

```
labels = ['Unique Open Rate', 'Open Rate', 'Click Rate', 'Conversions']
data_19 = [open_rate_19, open_rate_imodules_19, click_rate_19, conversions_19]
data_20 = [open_rate_20, open_rate_imodules_20, click_rate_20, conversions_20]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, data_19, width, label='2019')
rects2 = ax.bar(x + width/2, data_20, width, label='2020')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Comparions')
ax.set_title('Email Reporting 2019 vs 2020')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
